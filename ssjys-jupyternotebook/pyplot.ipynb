```python
#直线图
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(1,11)
y=2*x+5

plt.title("Matplotlib Demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
#圆点图，使用ob作为plot（）函数中的格式字符串
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(1,11)
y=2*x+5
plt.title("Matplotlib Demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,"ob")
plt.show()
```

```python
#绘制郑弦波
import numpy as np
from matplotlib import pyplot as plt

x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show()

```

```python
#pyplot子模块提供bar()函数生成条形图

from matplotlib import pyplot as plt
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]

plt.title("Bar graph")
plt.bar(x,y,align="center")
plt.bar(x2,y2,color='g',align="center")

plt.show()
```

```python
#添加图例legend
#当线条过多时，我们设置不同颜色来区分不同线条。因此，需要对不同颜色线条做下标注，我们实用 legend() 接口来实现
import numpy as np
import matplotlib.pyplot as plt

# 显示中文
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False

%matplotlib inline

x=np.arange(1,11)

plt.plot(x,x)
plt.plot(x,x*2)
plt.plot(x,x*3)
plt.plot(x,x*4)

plt.legend(['生活','颜值','工作','金钱'])

plt.show()
```

```python
#切换线条样式marker
#如果想改变线条的样式，我们可以使用修改 plot() 绘图接口中 mark 参数

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x=np.arange(1,5)

plt.plot(x,x,marker='o')
plt.plot(x,x+1,marker='>')
plt.plot(x,x+2,marker='s')
plt.show()
```

```python
#随机
import numpy as np
import matplotlib.pyplot as plt

x=np.random.normal(size=(2,100))
y1=np.random.normal(size=(2,100))
y2=np.random.normal(size=(2,100))

plt.plot(x,y1)
plt.plot(x,y2)

plt.show()
```
